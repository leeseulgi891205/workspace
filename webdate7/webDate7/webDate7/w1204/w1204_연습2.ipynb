```python
# 화면스크롤 내림
pre_height = browser.execute_script("return document.body.scrollHeight")
pre_height
```

```python
while True:
    # 스크롤 내림
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight);")
    # 추가데이터를 불러오는동안 대기
    time.sleep(3)
    curr_height = browser.execute_script("return document.body.scrollHeight")
    if curr_height == pre_height:
        break
    # 같이 않으면
    pre_height = curr_height
    print("높이변화 : ",curr_height)
    
print("스크롤 완료 : ",curr_height)
```

```python
# 웹스크래핑
soup = BeautifulSoup(browser.page_source, "lxml")
results = soup.find("div",{"class":"domestic_results__gp5WB"})
flights = results.find_all("div",{"class":"domestic_Flight__8bR_b"})
flights[0]
```

```python
# 항공사명
flights[0].find("b",{"class":"airline_name__0Tw5w"}).text.strip()
```

```python
# 출발시간
flights[0].find_all("b",{"class":"route_time__xWu7a"})[0].text.strip()
# 도착시간
flights[0].find_all("b",{"class":"route_time__xWu7a"})[1].text.strip()
# 가격 인트 타입
int(flights[0].find_all("i",{"class":"domestic_num__shOub"})[1].text.strip().replace(",",""))
```

```python
# 항공사명, 출발시간, 도착시간, 가격 출력
for flight in flights:
    print()
    flight.find("b",{"class":"airline_name_0Tw5w"}).text.strip()
    flight.find_all("b",{"class":"route_time__xWu7a"})[0].text.strip()
    flight.find_all("b",{"class":"route_time__xWu7a"})[1].text.strip()
    int(flight.find_all("i",{"class":"domestic_num_shOub"})[1].text.strip().replace("",""))
    print(""*50)
```
