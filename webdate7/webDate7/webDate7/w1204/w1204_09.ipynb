```python
# 4. 스크롤 하단 끝까지 내리기
pre_height = browser.execute_script("return document.body.scrollHeight")
pre_height

while True:
    # 스크롤 아래로 내림
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    # 추가데이터를 불러오는동안 대기
    time.sleep(3)
    curr_height = browser.execute_script("return document.body.scrollHeight")
    if curr_height == pre_height:
        break
    # 같지 않으면
    pre_height = curr_height
    print("높이 변화 : ",curr_height)

print("스크롤 완료 : ",curr_height)  
```

```python
# 6. 평점 3.5 이상, 평가수 100 이상, 가격 200000 이하 크롤링

soup = BeautifulSoup(browser.page_source, "lxml")
cards = soup.select("div.css-gvoll6")   # 여기어때 숙소 카드

print("\n===== (평점 3.5↑, 평가수 100↑, 가격 200000↓) =====\n")

for i, card in enumerate(cards):
    print(f"{i+1}.")

    # 1) 이미지
    img_tag = card.find("img")
    img_url = img_tag.get("data-src") or img_tag.get("src") if img_tag else ""

    # 2) 제목 — ★여기어때에서 가장 안정적인 방식 (강력 추천)
    title_tag = card.find("strong")
    title = title_tag.get_text(strip=True) if title_tag else "제목 없음"

    # 3) 가격
    price_tag = card.find("span", {"class": "css-5r5920"})
    if not price_tag:
        continue
    price_text = price_tag.get_text(strip=True)
    price = int(price_text.replace(",", "").replace("원", ""))

    # 4) 평점
    rating_tag = card.find("span", {"class": "css-9ml4lz"})
    if not rating_tag:
        continue
    rating = float(rating_tag.get_text(strip=True))

    # 5) 평가수 (예: "1,284 명 평가" → 1284)
    ratingCount_tag = card.find("span", {"class": "css-oj6onp"})
    if not ratingCount_tag:
        continue
    ratingCount_text = ratingCount_tag.get_text(strip=True)
    ratingCount = int("".join(ch for ch in ratingCount_text if ch.isdigit()))

    # 6) 조건 필터
    if rating >= 3.5 and ratingCount >= 100 and price <= 200000:
        print(f"제목   : {title}")
        print(f"이미지 : {img_url}")
        print(f"평점   : {rating}")
        print(f"평가수 : {ratingCount}")
        print(f"가격   : {price}")
        print("-" * 50)
    else:
        continue
```

```python
# 6. csv파일로 저장하시오.
csv_filename = "yeogi_gangneung_hotels111.csv"
soup = BeautifulSoup(browser.page_source, "lxml")
items = soup.select("div.css-gvoll6")

with open(csv_filename, mode="w", encoding="utf-8", newline="") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["이미지", "제목", "평점", "평가수", "가격"])
    for item in items:
        # 이미지
        img_tag = item.select_one("img.thumbnail-image")
        img_url = img_tag.get("data-src") or img_tag.get("src") or "" if img_tag else ""
        # 제목
        title_tag = item.select_one(".css-1gxx2ac")
        title = title_tag.get_text(strip=True) if title_tag else ""
        # 평점
        rating_tag = item.select_one(".css-9ml4lz")
        rating = rating_tag.get_text(strip=True) if rating_tag else ""
        # 평가수
        count_tag = item.select_one(".css-oj6onp")
        count = count_tag.get_text(strip=True) if count_tag else ""
        # 가격
        price_tag = item.select_one(".css-5r5920")
        price = price_tag.get_text(strip=True) if price_tag else ""
        writer.writerow([img_url, title, rating, count, price])

print(f"데이터가 저장되었습니다.")
```
