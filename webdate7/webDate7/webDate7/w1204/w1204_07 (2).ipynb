```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

import requests
from bs4 import BeautifulSoup
import time
import os
import random
import csv
```

```python

url = "https://flight.naver.com/"
browser = webdriver.Chrome() # 창열기
browser.maximize_window() # 최대창 확대
# 1. naver 페이지 열기
browser.get(url)

# 다른 탭 선택 - 네이버항공권 탭 선택됨.
# browser.switch_to.window(browser.window_handles[1])
```

```python
# 광고알림창 닫기 버튼
elem = browser.find_element(By.XPATH,'//*[@id="layer"]/button[2]')
time.sleep(1)
elem.click()
```

```python
# 1. 김포, 제주 선택
# 김포선택
browser.find_element(By.XPATH,'//*[@id="__next"]/div/main/div[2]/div/div/div[2]/div[1]/button[1]').click()
time.sleep(1)
browser.find_element(By.XPATH,'//*[@id="__next"]/div/main/div[8]/div[2]/div/div/ul[1]/li[3]/button').click()
# 제주선택
browser.find_element(By.XPATH,'//*[@id="__next"]/div/main/div[2]/div/div/div[2]/div[1]/button[2]').click()
time.sleep(1)
browser.find_element(By.XPATH,'//*[@id="__next"]/div/main/div[8]/div[2]/div[2]/div[2]/ul[1]/li[1]/button').click()


```

```python
# 날짜선택 - 달력열기
elem = WebDriverWait(browser, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div/main/div[2]/div/div/div[2]/div[2]/button[1]')))

time.sleep(2)

# 가는날선택
elem = WebDriverWait(browser, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div/main/div[8]/div[2]/div[1]/div[2]/div/div/div[2]/table/tbody/tr[5]/td[2]/button')))
elem.click()
time.sleep(1)

# 오는날선택
elem = WebDriverWait(browser, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div/main/div[8]/div[2]/div[1]/div[2]/div/div/div[2]/table/tbody/tr[5]/td[3]/button')))

elem.click()
time.sleep(1)

# 검색선택
elem = WebDriverWait(browser, 10).until(
    EC.element_to_be_clickable((By.XPATH, '//*[@id="__next"]/div/main/div[2]/div/div/div[2]/div[4]/button')))
elem.click()
time.sleep(2)
```

```python
# 화면이 로딩될때까지 대기
WebDriverWait(browser, 10).until(
    EC.presence_of_element_located((By.XPATH, '//*[@id="__next"]/div/main/div[4]/div/div[2]/div[2]/div[1]'))
)
```

```python
# 화면스크롤 내림
pre_height = browser.execute_script("return document.body.scrollHeight")
pre_height
```

```python
while True:
    # 스크롤 아래로 내림
    browser.execute_script("window.scrollTo(0,document.body.scrollHeight)")
    # 추가데이터를 불러오는동안 대기
    time.sleep(3)
    curr_height = browser.execute_script("return document.body.scrollHeight")
    if curr_height == pre_height:
        break
    # 같지 않으면
    pre_height = curr_height
    print("높이 변화 : ",curr_height)

print("스크롤 완료 : ",curr_height)    
    
    
    
```

```python
soup = BeautifulSoup(browser.page_source,"lxml")
results = soup.find("div",{"class":"domestic_results__gp5WB"})
flights = results.find_all("div",{"class":"domestic_Flight__8bR_b"})
flights[0]
```

```python
# csv파일로 저장하시오.
# 엑셀파일로 저장
# 항공사명,출발시간,도착시간,가격
# 파라타항공,06:05,07:20,26700
f = open("flight.csv","w",encoding="utf-8",newline="")
writer = csv.writer(f)
title = ['항공사명\t','출발시간\t','도착시간\t','가격\t']
writer.writerow(title)
print('-'*50)
```

```python
# 엑셀파일로 저장
for i,flight in enumerate(flights):
    print(f"{i+1}.",end="\t")
    # 항공사명
    print(flight.find("b",{"class":"airline_name__0Tw5w"}).text.strip(),end="\t")
    print("",end="\t")
    # 항공사명 저장
    n1 = flight.find("b",{"class":"airline_name__0Tw5w"}).text.strip()
    # 출발시간/도착시간
    print(flight.find_all("b",{"class":"route_time__xWu7a"})[0].text.strip(),end="\t")
    print("",end="\t")
    # 출발시간 저장
    n2 = flight.find_all("b",{"class":"route_time__xWu7a"})[0].text.strip()
    # 도착시간 저장
    n3 = flight.find_all("b",{"class":"route_time__xWu7a"})[1].text.strip()
    print(flight.find_all("b",{"class":"route_time__xWu7a"})[1].text.strip(),end="\t")
    print("",end="\t")
    # 가격 저장
    n4 = int(flight.find("i",{"class":"domestic_num__ShOub"}).text.strip().replace(",",""))
    print(int(flight.find("i",{"class":"domestic_num__ShOub"}).text.strip().replace(",","")))
    print("-"*50)
```

```python
# 항공사명,출발시간,도착시간,가격 출력
for i, flight in enumerate(flights):
    try:
        print(f"{i+1}.", end="\t")
        
        # 항공사명
        n1 = flight.find("b", {"class": "airline_name__0Tw5w"}).text.strip()
        print(n1, end="\t")
        
        # 출발시간/도착시간
        times = flight.find_all("b", {"class": "route_time__xWu7a"})
        n2 = times[0].text.strip()  # 출발시간
        n3 = times[1].text.strip()  # 도착시간
        print(n2, end="\t")
        print(n3, end="\t")
        
        # 가격
        n4 = int(flight.find("i", {"class": "domestic_num__ShOub"}).text.strip().replace(",", ""))
        print(n4)
        print("-" * 50)
        
        # CSV 파일에 작성
        arr2 = [n1, n2, n3, n4]
        writer.writerow(arr2)
        
    except Exception as e:
        print(f"항공편 {i+1} 처리 중 오류: {e}")
        continue
    
# 파일 닫기
f.close()
print("=" * 50)
print("flight.csv 파일 저장 완료!")
```
