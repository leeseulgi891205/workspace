```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import requests
from bs4 import BeautifulSoup
import time
import os
import random
import csv
```

```python
url = "https://www.melon.com/"
browser = webdriver.Chrome() # 창열기
browser.maximize_window() # 최대창 확대
# 1. naver 페이지 열기
browser.get(url)
```

```python
# 광고알림창 닫기 버튼
elem = browser.find_element(By.XPATH,'//*[@id="mainPop"]/div/div[2]/div[2]/button/span/span')
time.sleep(1)
elem.click()
```

```python
# 멜론차트 선택
browser.find_element(By.XPATH,'//*[@id="gnb_menu"]/ul[1]/li[1]/a/span[2]').click()
time.sleep(2)
```

```python
# 멜론 차트 순위 곡정보 앨범 좋아요 출력
soup = BeautifulSoup(browser.page_source, "lxml")
frm = soup.find("form", {"id": "frm"})
trs = frm.find("tbody").find_all("tr")
seen = set()

for tr in trs:
    title = tr.find("div", {"class": "ellipsis rank01"}).get_text(strip=True)
    
    if title not in seen:
        seen.add(title)
        rank = tr.find("span", {"class": "rank"}).get_text(strip=True)
        artist = tr.find("div", {"class": "ellipsis rank02"}).get_text(strip=True)
        album = tr.find("div", {"class": "ellipsis rank03"}).get_text(strip=True)
        like = tr.find("span", {"class": "cnt"}).get_text(strip=True)
        print(rank, title, artist, album, like)
```

```python
# 멜론 좋아요 100,000 이상 곡정보 출력
soup = BeautifulSoup(browser.page_source, "lxml")
frm = soup.find("form", {"id": "frm"})
trs = frm.find("tbody").find_all("tr")

for tr in trs:
    rank = tr.find("span", {"class": "rank"}).get_text(strip=True)
    title = tr.find("div", {"class": "ellipsis rank01"}).get_text(strip=True)
    artist = tr.find("div", {"class": "ellipsis rank02"}).get_text(strip=True)
    album = tr.find("div", {"class": "ellipsis rank03"}).get_text(strip=True)
    like_tag = tr.find("span", {"class": "cnt"})
    if not like_tag:
        continue
    like = like_tag.get_text(strip=True).replace(",", "")
    like_num = ''.join(filter(str.isdigit, like))
    if like_num and int(like_num) >= 100000:
        print(rank, title, artist, album, like_num)
```

```python
# 멜론차트 이미지까지 출력
soup = BeautifulSoup(browser.page_source, "lxml")
frm = soup.find("form", {"id": "frm"})
trs = frm.find("tbody").find_all("tr")
for tr in trs:
    rank = tr.find("span", {"class": "rank"}).get_text(strip=True)
    title = tr.find("div", {"class": "ellipsis rank01"}).get_text(strip=True)
    artist = tr.find("div", {"class": "ellipsis rank02"}).get_text(strip=True)
    album = tr.find("div", {"class": "ellipsis rank03"}).get_text(strip=True)
    like = tr.find("span", {"class": "cnt"}).get_text(strip=True)
    img_tag = tr.find("img")
    img_url = img_tag['src'] if img_tag else 'No Image'
    print(rank, title, artist, album, like, img_url)
browser.quit()
```

```python
# 멜론차트 70등만 출력
soup = BeautifulSoup(browser.page_source,"lxml")
frm = soup.find("form",{"id":"frm"})
trs = frm.find("tbody").find_all("tr")
for tr in trs:
    rank = tr.find("span",{"class":"rank"}).get_text()
    if rank == '70':
        title = tr.find("div",{"class":"ellipsis rank01"}).get_text().strip()
        artist = tr.find("div",{"class":"ellipsis rank02"}).get_text().strip()
        album = tr.find("div",{"class":"ellipsis rank03"}).get_text().strip()
        like = tr.find("span",{"class":"cnt"}).get_text().strip()
        print(rank,title,artist,album,like)
```

```python
# csv 파일로 저장
soup = BeautifulSoup(browser.page_source, "lxml")
frm = soup.find("form", {"id": "frm"})
trs = frm.find("tbody").find_all("tr")

with open("melon.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.writer(f)
    writer.writerow(["순위", "곡명", "아티스트", "앨범", "좋아요"])
    
    seen = set()
    for tr in trs:
        title = tr.find("div", {"class": "ellipsis rank01"}).get_text(strip=True)
        
        if title not in seen:
            seen.add(title)
            rank = tr.find("span", {"class": "rank"}).get_text(strip=True)
            artist = tr.find("div", {"class": "ellipsis rank02"}).get_text(strip=True)
            album = tr.find("div", {"class": "ellipsis rank03"}).get_text(strip=True)
            like_tag = tr.find("span", {"class": "cnt"})
            like = like_tag.get_text(strip=True) if like_tag else "0"
            writer.writerow([rank, title, artist, album, like])

print("melon.csv 파일이 저장되었습니다.")
```

```python
# 멜론차트 hot100 클릭
browser.find_element(By.XPATH,'//*[@id="gnb_menu"]/ul[1]/li[1]/div/ul/li[2]/a/span').click()
time.sleep(2)
```

```python
# 멜론차트 hot100 스크롤 하단까지 내리기
pre_height = browser.execute_script("return document.body.scrollHeight")

while True:
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
    time.sleep(2)
    curr_height = browser.execute_script("return document.body.scrollHeight")
    if curr_height == pre_height:
        break
    pre_height = curr_height

print("스크롤 완료")

# 멜론차트 hot100 순위 곡정보 앨범 좋아요 전체 출력
soup = BeautifulSoup(browser.page_source, "lxml")
frm = soup.find("form", {"id": "frm"})
trs = frm.find("tbody").find_all("tr")

print(f"\n총 {len(trs)}곡\n")

seen = set()
for tr in trs:
    rank = tr.find("span", {"class": "rank"}).get_text(strip=True)
    title = tr.find("div", {"class": "ellipsis rank01"}).get_text(strip=True)
    
    # 중복 체크
    if title in seen:
        continue
    seen.add(title)
    
    artist = tr.find("div", {"class": "ellipsis rank02"}).get_text(strip=True)
    album = tr.find("div", {"class": "ellipsis rank03"}).get_text(strip=True)
    like_tag = tr.find("span", {"class": "cnt"})
    like = like_tag.get_text(strip=True) if like_tag else "0"
    
    print(rank, title, artist, album, like)
```
