```python
for i in range(1,6):
    print(f"{i}")
    url = f"https://www.gmarket.co.kr/n/search?keyword=tv&p={i}"
    print(url)
```

```python
# 후기 50개이상 구매 100건 이상 가격 50만원 이상 스크래핑.
driver = webdriver.Chrome()
driver.implicitly_wait(5)
driver.get(url)
time.sleep(2)
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
items = soup.select('.box__item-container')

for item in items:
    name = item.select_one('.text__item').get_text().strip()
    price = item.select_one('.text__value').get_text().strip()
    print(name, price)  
```

```python
from bs4 import BeautifulSoup
import re

soup = BeautifulSoup(html, 'html.parser')
items = soup.select('.box__item-container')

def get_num(text):
    m = re.search(r'(\d[\d,]*)', text or "")
    return m.group(1).replace(",", "") if m else "0"

for item in items:
    img_tag = item.select_one("img")
    image = img_tag.get("src") if img_tag else "이미지 없음"

    name_tag = item.select_one(".text__item")
    name = name_tag.get_text(strip=True) if name_tag else ""

    price_tag = item.select_one(".text__value")
    price = int(get_num(price_tag.get_text())) if price_tag else 0

    rating_tag = item.select_one(".for-a11y")
    rating = get_num(rating_tag.get_text()) if rating_tag else "0"

    review_tag = item.select_one(".list-item__feedback-count")
    review_count = get_num(review_tag.get_text()) if review_tag else "0"

    purchase_tag = item.select_one(".list-item__pay-count")
    purchase_count = get_num(purchase_tag.get_text()) if purchase_tag else "0"

    print(f"이미지: {image}")
    print(f"제목: {name}")
    print(f"가격: {price:,}원")
    print(f"평점: {rating}%")
    print(f"후기: {review_count}")
    print(f"구매: {purchase_count}")
    print("-" * 80)
```
